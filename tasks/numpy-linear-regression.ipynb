```python
import numpy as np
import matplotlib.pyplot as plt

# expected weights = 0
x_train = np.linspace(0, 5, 100)
y_train = x_train + (np.random.randn(100) * 0.3)
```

```python
W = np.random.rand()
hypothesis = x_train * W
_past_hypothesis = hypothesis

# 트레이닝 데이터와 초기 데이터 보여주기
plt.scatter(x_train, y_train)
plt.plot(x_train, hypothesis, color='red')
plt.show()
```

```python
for step in range(0, 2000):
    hypothesis = x_train * W
    cost = np.average(np.square(hypothesis - y_train))
    gradient = np.average((hypothesis - y_train) * x_train)
    
    if step % 50 == 0:
        print("cost: %s, W: %s" % (cost, W))
    
    W = W - 0.001 * gradient
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, _past_hypothesis)
plt.plot(x_train, hypothesis, color='red')
plt.show()
```
